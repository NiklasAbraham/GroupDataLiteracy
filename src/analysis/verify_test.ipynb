```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# import PCA, TSNE, UMAP
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from umap import UMAP
```

```python
# data/movie_embeddings.npy
# /home/nab/GroupDataLiteracy/data/movie_embeddings.npy
embeddings = np.load('/home/nab/GroupDataLiteracy/data/movie_embeddings.npy')
```

```python
# give me PCA of the embeddings
pca = PCA(n_components=2)
pca.fit(embeddings)
embeddings_pca = pca.transform(embeddings)

# plot the PCA
plt.scatter(embeddings_pca[:, 0], embeddings_pca[:, 1])
plt.show()
```

```python
# give me t-SNE of the embeddings
tsne = TSNE(n_components=2, random_state=42)
embeddings_tsne = tsne.fit_transform(embeddings)

# plot the t-SNE
plt.scatter(embeddings_tsne[:, 0], embeddings_tsne[:, 1])
plt.title('t-SNE Visualization of Movie Embeddings')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()


# give me UMAP of the embeddings
umap = UMAP(n_components=2, random_state=42)
embeddings_umap = umap.fit_transform(embeddings)

# plot the UMAP
plt.scatter(embeddings_umap[:, 0], embeddings_umap[:, 1])
plt.title('UMAP Visualization of Movie Embeddings')
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.show()
```
